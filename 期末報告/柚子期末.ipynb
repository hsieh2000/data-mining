```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import random
```

```python
dataset = pd.read_csv('cardio_train.csv', sep=';', index_col=0)
dataset['height1']=dataset['height']/100
dataset['BMI']=round((dataset['weight']/(dataset['height1']**2)),2)
dataset['avg_ap']=round((dataset['ap_hi']+2*dataset['ap_lo'])/3,4)
dataset
```

```python
dataset=dataset.drop(dataset[dataset['BMI']>50].index)
dataset=dataset.drop(dataset[dataset['BMI']<15].index)
dataset
```

```python
dataset=dataset.drop(dataset[dataset['avg_ap']>110].index)
dataset=dataset.drop(dataset[dataset['avg_ap']<70].index)
dataset
```

```python
sub_dataset=dataset.iloc[ :, [-1,-2]]
sub_dataset
```

```python
dataset
```

```python
X=sub_dataset.iloc[:,[-1,-2]].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
y_kmeans=y_kmeans.tolist()
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='green',label='Cluster 3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=100,c='magenta',label='Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],s=300,c='yellow',label='Centroids')
plt.title('Clusters of customers')
plt.xlabel('avg_ap')
plt.ylabel('BMI')
plt.legend()
plt.show()
```

```python
dataset['clustering']=y_kmeans
```
