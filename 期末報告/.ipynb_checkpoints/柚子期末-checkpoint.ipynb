```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('cardio_train.csv', sep=';', index_col=0)
#df=df1[0:100]
df['height1']=df['height']/100
df['BMI']=round((df['weight']/(df['height1']**2)),2)
df['avg_ap']=round((df['ap_hi']+2*df['ap_lo'])/3,4)
df
```

```python
df1=df.drop(df[df['BMI']>50].index)
df2=df1.drop(df1[df1['BMI']<15].index)
df2
```

```python
df3=df2.drop(df2[df2['avg_ap']>110].index)
df4=df3.drop(df3[df3['avg_ap']<70].index)
df4
```

```python
sub_dataset=df4.iloc[ :, [-1,-2]]
sub_dataset
```

```python
X=sub_dataset.iloc[:,[0,1]].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
len(y_kmeans)
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='green',label='Cluster 3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=100,c='magenta',label='Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],s=300,c='yellow',label='Centroids')
plt.title('Clusters of customers')
plt.xlabel('avg_ap')
plt.ylabel('BMI')
plt.legend()
plt.show()
```
